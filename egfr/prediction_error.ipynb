```julia
augmented_run = "204"
baseline_run = "205"
for run_num in [baseline_run, augmented_run]
    saveat = deserialize("outputs/000_processed_grb_egfr_20.dict")["save_at"].+1 #consistent across all, need to add 1 for correct indexing
    n_test_points = length(saveat)
    dose_names = ["20","2","02"] #all possible doses 
    quantity_names = ["p_egfr", "p_shc", "p_plcg", "grb_egfr", "grb_shc"] #all possible species
    PPP_string = ["\nLog Posterior Predictive Probability\n"]
    for i in quantity_names
        if i == "p_egfr"
            dose_names = ["20","2","02"]
            for j in dose_names
                test_predictions = deserialize("outputs/500_$(run_num)_predictions_$(i)_$(j).jls")
                n_predictions = length(test_predictions)
                test_data = deserialize("outputs/000_processed_$(i)_$(j).dict")["response"]
                test_data_error = deserialize("outputs/000_processed_$(i)_$(j).dict")["average_error"]
                log_likelihood = []
                #note - we take individual log likelihoods, rather than joint, because the joint is less numerically stable
                for i in 1:n_predictions
                    push!(log_likelihood, [loglikelihood(Normal(test_predictions[i][j], test_data_error), test_data[j]) for j in 1:n_test_points]);
                end
                    #Bootstrap Likelihood Samples 500 Times
                rng = StableRNG(394948)
                n_bootstraps = 500
                indices = [rand(rng, 1:n_predictions, n_predictions) for i in 1:n_bootstraps]
                bootstrapped_likelihood = [log_likelihood[indices[i]] for i in 1:n_bootstraps];

                #Calculate average log posterior predictive probability of test data
                log_sum_exp_bootstrapped = [vec(logsumexp(Base.stack(bootstrapped_likelihood[i]), dims=2).-log(n_predictions)) for i in 1:n_bootstraps]
                log_posterior_predictive_probability_bootstrapped = vec(sum(Base.stack(log_sum_exp_bootstrapped),dims=1))
                
                text_out = "$(i)"*" "*"$(j): "* "$(mean(log_posterior_predictive_probability_bootstrapped)) +/- $(sqrt(var(log_posterior_predictive_probability_bootstrapped)))\n"
                push!(PPP_string, text_out)
            end
        end
        if i == "p_shc" 
            dose_names = ["20"]
            for j in dose_names
                test_predictions = deserialize("outputs/500_$(run_num)_predictions_$(i)_$(j).jls")
                n_predictions = length(test_predictions)
                test_data = deserialize("outputs/000_processed_$(i)_$(j).dict")["response"]
                test_data_error = deserialize("outputs/000_processed_$(i)_$(j).dict")["average_error"]
                log_likelihood = []
                #note - we take individual log likelihoods, rather than joint, because the joint is less numerically stable
                for i in 1:n_predictions
                    push!(log_likelihood, [loglikelihood(Normal(test_predictions[i][j], test_data_error), test_data[j]) for j in 1:n_test_points]);
                end
                    #Bootstrap Likelihood Samples 500 Times
                rng = StableRNG(394948)
                n_bootstraps = 500
                indices = [rand(rng, 1:n_predictions, n_predictions) for i in 1:n_bootstraps]
                bootstrapped_likelihood = [log_likelihood[indices[i]] for i in 1:n_bootstraps];

                #Calculate average log posterior predictive probability of test data
                log_sum_exp_bootstrapped = [vec(logsumexp(Base.stack(bootstrapped_likelihood[i]), dims=2).-log(n_predictions)) for i in 1:n_bootstraps]
                log_posterior_predictive_probability_bootstrapped = vec(sum(Base.stack(log_sum_exp_bootstrapped),dims=1))
                
                text_out = "$(i)"*" "*"$(j): "* "$(mean(log_posterior_predictive_probability_bootstrapped)) +/- $(sqrt(var(log_posterior_predictive_probability_bootstrapped)))\n"
                push!(PPP_string, text_out)
            end
        end
        if i == "p_plcg"
            dose_names = ["20","2"]
            for j in dose_names
                test_predictions = deserialize("outputs/500_$(run_num)_predictions_$(i)_$(j).jls")
                n_predictions = length(test_predictions)
                test_data = deserialize("outputs/000_processed_$(i)_$(j).dict")["response"]
                test_data_error = deserialize("outputs/000_processed_$(i)_$(j).dict")["average_error"]
                log_likelihood = []
                #note - we take individual log likelihoods, rather than joint, because the joint is less numerically stable
                for i in 1:n_predictions
                    push!(log_likelihood, [loglikelihood(Normal(test_predictions[i][j], test_data_error), test_data[j]) for j in 1:n_test_points]);
                end
                    #Bootstrap Likelihood Samples 500 Times
                rng = StableRNG(394948)
                n_bootstraps = 500
                indices = [rand(rng, 1:n_predictions, n_predictions) for i in 1:n_bootstraps]
                bootstrapped_likelihood = [log_likelihood[indices[i]] for i in 1:n_bootstraps];

                #Calculate average log posterior predictive probability of test data
                log_sum_exp_bootstrapped = [vec(logsumexp(Base.stack(bootstrapped_likelihood[i]), dims=2).-log(n_predictions)) for i in 1:n_bootstraps]
                log_posterior_predictive_probability_bootstrapped = vec(sum(Base.stack(log_sum_exp_bootstrapped),dims=1))
                
                text_out = "$(i)"*" "*"$(j): "* "$(mean(log_posterior_predictive_probability_bootstrapped)) +/- $(sqrt(var(log_posterior_predictive_probability_bootstrapped)))\n"
                push!(PPP_string, text_out)
            end
        end
        if i == "grb_egfr"
            dose_names = ["20"]
            for j in dose_names
                test_predictions = deserialize("outputs/500_$(run_num)_predictions_$(i)_$(j).jls")
                n_predictions = length(test_predictions)
                test_data = deserialize("outputs/000_processed_$(i)_$(j).dict")["response"]
                test_data_error = deserialize("outputs/000_processed_$(i)_$(j).dict")["average_error"]
                log_likelihood = []
                #note - we take individual log likelihoods, rather than joint, because the joint is less numerically stable
                for i in 1:n_predictions
                    push!(log_likelihood, [loglikelihood(Normal(test_predictions[i][j], test_data_error), test_data[j]) for j in 1:n_test_points]);
                end
                    #Bootstrap Likelihood Samples 500 Times
                rng = StableRNG(394948)
                n_bootstraps = 500
                indices = [rand(rng, 1:n_predictions, n_predictions) for i in 1:n_bootstraps]
                bootstrapped_likelihood = [log_likelihood[indices[i]] for i in 1:n_bootstraps];

                #Calculate average log posterior predictive probability of test data
                log_sum_exp_bootstrapped = [vec(logsumexp(Base.stack(bootstrapped_likelihood[i]), dims=2).-log(n_predictions)) for i in 1:n_bootstraps]
                log_posterior_predictive_probability_bootstrapped = vec(sum(Base.stack(log_sum_exp_bootstrapped),dims=1))
                
                text_out = "$(i)"*" "*"$(j): "* "$(mean(log_posterior_predictive_probability_bootstrapped)) +/- $(sqrt(var(log_posterior_predictive_probability_bootstrapped)))\n"
                push!(PPP_string, text_out)
            end
        end
        if i == "grb_shc"
            dose_names = ["20"]
            for j in dose_names
                test_predictions = deserialize("outputs/500_$(run_num)_predictions_$(i)_$(j).jls")
                n_predictions = length(test_predictions)
                test_data = deserialize("outputs/000_processed_$(i)_$(j).dict")["response"]
                test_data_error = deserialize("outputs/000_processed_$(i)_$(j).dict")["average_error"]
                log_likelihood = []
                #note - we take individual log likelihoods, rather than joint, because the joint is less numerically stable
                for i in 1:n_predictions
                    push!(log_likelihood, [loglikelihood(Normal(test_predictions[i][j], test_data_error), test_data[j]) for j in 1:n_test_points]);
                end
                    #Bootstrap Likelihood Samples 500 Times
                rng = StableRNG(394948)
                n_bootstraps = 500
                indices = [rand(rng, 1:n_predictions, n_predictions) for i in 1:n_bootstraps]
                bootstrapped_likelihood = [log_likelihood[indices[i]] for i in 1:n_bootstraps];

                #Calculate average log posterior predictive probability of test data
                log_sum_exp_bootstrapped = [vec(logsumexp(Base.stack(bootstrapped_likelihood[i]), dims=2).-log(n_predictions)) for i in 1:n_bootstraps]
                log_posterior_predictive_probability_bootstrapped = vec(sum(Base.stack(log_sum_exp_bootstrapped),dims=1))
                
                text_out = "$(i)"*" "*"$(j): "* "$(mean(log_posterior_predictive_probability_bootstrapped)) +/- $(sqrt(var(log_posterior_predictive_probability_bootstrapped)))\n"
                push!(PPP_string, text_out)
            end
        end
    end
    serialize("outputs/600_$(run_num)_posterior_predictive_prob.txt", join(PPP_string))
end
```
